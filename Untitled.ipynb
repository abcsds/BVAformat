```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pn1 = pd.read_csv("example/PeaksN1.txt", skiprows=1, index_col=0, delim_whitespace=True)
pn1.head()
```

```python
electrodes = ["Fp1", "AF3", "F7",  "F3",  "FC1", "FC5", "T7",  "C3",  "CP1", "CP5", "P7",  "P3",  "Pz",  "PO3", "O1",  "Oz",  "O2",  "PO4", "P4",  "P8",  "CP6", "CP2", "C4",  "T8",  "FC6", "FC2", "F4",  "F8",  "AF4", "Fp2", "Fz",  "Cz"]
```

```python
pn1.stack()
```

```python
mat = pn1.values
mat[0]
```

```python
mat = pn1.values
columns = ["Electrode", "Time", "Voltage"]
index = "VP"
# df = pd.DataFrame(index=index, columns=columns)
arr = np.empty((0, 4))
for vp, row in enumerate(mat):
    for i, t in enumerate(row):
        if i%2 == 0:
            new = np.array([vp+1, electrodes[int(i/2)],t, row[int(i/2+1)]])
            arr = np.append(arr, new.reshape(-1,4),
                           axis=0)
arr
```

```python
mat = pn1.values
columns = ["Electrode", "Time", "Voltage"]
index = "VP"
arr = np.empty((0, 4))
for vp, row in enumerate(mat):
    for i, t in enumerate(row):
        if i%2 == 0:
            new = np.array([vp+1,
                            electrodes[int(i/2)],
                            round(t, 5),
                            round(row[i+1], 5)])
            arr = np.append(arr, new.reshape(-1,4),
                           axis=0)
df = pd.DataFrame(arr[:,1:], index=arr[:,0], columns=columns)
df.index.name = index
df.head(10)
```

```python
df.to_csv("PN1.csv")
```

```python
mn1 = pd.read_csv("example/MeanN1.txt", index_col=0, delim_whitespace=True)
conditions = ["MA","AO","MV","VO","MVA","VAO"]
electrodes = ["Fz", "Cz"]
mn1.head()
```

```python
mat = mn1.values
mat
```

```python
columns = ["Channel", "Condition", "Mean_Voltage"]
index = "VP"
```

```python
arr = np.empty((0, 4))
for vp, row in enumerate(mat):
    for i, cond in enumerate(row):
        new = np.array([vp+1,
                        electrodes[i//len(conditions)],
                        conditions[i%len(conditions)],
                        round(cond, 5)])
        arr = np.append(arr, new.reshape(-1,4),
                        axis=0)
df = pd.DataFrame(arr[:,1:], index=arr[:,0], columns=columns)
df.index.name = index
df.head()
```

```python
df.to_csv('MN1.csv')
print(df.shape)
!head MN1.csv
```
